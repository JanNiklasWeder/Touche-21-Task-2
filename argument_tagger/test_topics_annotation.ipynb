```python
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
import xml.etree.ElementTree as ET
import pandas as pd

n_topics=50

def get_titles(file):
    tree = ET.parse(file)
    root = tree.getroot()
    buffer = []
    i=1
    for title in root.iter('title'):
        if i<=int(n_topics):
            title = title.text.strip()
            if notation_comparative_topic(title) == True:
                buffer.append((title, True))
            else:
                buffer.append((title,False))
        i=i+1
    return buffer
def notation_comparative_topic(text):
    doc = nlp(text)
    for token in doc:
        if token.tag_=="JJR" or token.tag_=="RBR": 
            '''
            check comparative sentences (not include superalive sentences)
            comparative sentences: comparative adjects, comparative adverbs
            '''
            return 1
        if token.text == "or" or token.lemma_ == "prefer": #when verbs, ex: prefer, using "or" for recognation?!
            return 1
    return 0
```

```python
text="What are the advantages and disadvantages of PHP over Python and vice versa?"
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.label_)
'''
result is not expected, because only Python als NEG recognized
'''
```
