```python
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
import xml.etree.ElementTree as ET
import pandas as pd

n_topics=50

def get_titles(file):
    tree = ET.parse(file)
    root = tree.getroot()
    relevant_annotations = []
    annotations = []
    reducing_annotations=[]
    i=1
    for title in root.iter('title'):
        if i<=int(n_topics):
            title = title.text.strip()
            relevant_annotations.append(get_relevant_annotations(title))
            reducing_annotations.append(get_reducing_annotations(title))
            annotations.append((title,get_annotations(title)))
        i=i+1
    return annotations, relevant_annotations, reducing_annotations

def get_annotations(text):
    doc = nlp(text)
    l = []
    for token in doc:
        l.append((token.text, token.tag_))
    return l
def get_relevant_annotations(data):
    nouns_as_string=[]
    doc = nlp(data)
    annotations = ['CC','CD',
                    'JJ','JJR','JJS',
                    'RB','RBR','RBS', 
                    'NN', 'NNS','NNP','NNPS', 
                    'VB']
    for token in doc:
        if token.tag_ in annotations:
            nouns_as_string.append(token.text)
    return ' '.join(nouns_as_string)
def get_reducing_annotations(data):
    nouns_as_string=[]
    doc = nlp(data)
    annotations = ['CC','CD',
                    'JJS',
                    'RB','RBR','RBS', 
                    'NN', 'NNS','NNP','NNPS', 
                    'VB']
    for token in doc:
        if token.tag_ in annotations:
            nouns_as_string.append(token.text)
    return ' '.join(nouns_as_string)
```

```python
annotations, annotations_relevant, reduced_annotations = get_titles("topics-task-2.xml")

annotations_titles = [ annotations[i][0] for i in range(0,len(annotations))]
annotations_tags = [ annotations[i][1] for i in range(0,len(annotations))]

df = pd.DataFrame({"title": annotations_titles, "POS":annotations_tags, "rel_POS": annotations_relevant, "reduced":reduced_annotations})
df.index +=1
df
```
