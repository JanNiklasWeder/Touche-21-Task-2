```python
import requests
import xml.etree.ElementTree as ET
import time
from collections import Counter
from itertools import chain
import re
import pandas as pd
import numpy as np
import os

n_topics=50
def get_titles(file):
    tree = ET.parse(file)
    root = tree.getroot()
    buffer = []
    i=1
    for title in root.iter('title'):
        if i<=int(n_topics):
            buffer.append(title.text.strip())
        i=i+1
    return buffer
```
